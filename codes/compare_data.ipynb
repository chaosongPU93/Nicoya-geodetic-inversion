```python
import pandas as pd
import numpy as np

datadir = "/home/staff/chao/SSEinv/Nicoya/data/"

ckfig6 = "CKfig6_data.csv"
df_ck = pd.read_csv(datadir + ckfig6, sep=",", skiprows=1, \
                   names=['lon', 'lat', 'vx_Car', 'vy_Car', 'vz_Car'])

ckfig6_cs = "CKfig6_data_chao.csv"
df_cs = pd.read_csv(datadir + ckfig6_cs, sep=",", skiprows=1, \
                   names=['lon', 'lat', 'vx_Car', 'vy_Car', 'vz_Car', \
                          'vx_std_Car', 'vy_std_Car', 'vz_std_Car'])

# given station order is the same, compute difference 
# compute differences
df_diff = pd.DataFrame({
    'vx_diff': df_ck['vx_Car'] - df_cs['vx_Car'],
    'vy_diff': df_ck['vy_Car'] - df_cs['vy_Car'],
    'vz_diff': df_ck['vz_Car'] - df_cs['vz_Car']
})

# (optional) keep lon/lat if they exist in df1
if 'lon' in df_ck.columns and 'lat' in df_ck.columns:
    df_diff.insert(0, 'lon', df_ck['lon'])
    df_diff.insert(1, 'lat', df_ck['lat'])

print(df_diff.head())

print(df_diff['vx_diff'].min(), df_diff['vx_diff'].max())
print(df_diff['vy_diff'].min(), df_diff['vy_diff'].max())
print(df_diff['vz_diff'].min(), df_diff['vz_diff'].max())
```
